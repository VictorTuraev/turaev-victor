```python
class Vector():
    def __init__(self, x, y, z):
        self.x = x
        self.y = y
        self.z = z
    def __str__(self):
        return "<{}, {}, {}>".format( self.x, self.y, self.z)
    def add(self, other):
        return Vector(self.x+other.x, self.y+other.y, self.z+other.z)
    def __add__(self, other):
        return Vector(self.x+other.x, self.y+other.y, self.z+other.z)
    def __matmul__(self, other):
        return Vector(self.y*other.z - self.z*other.y, -self.x*other.z + self.z*other.x, self.x*other.y - self.y*other.x)
    def __mul__(self, other):
        return self.x*other.x + self.y*other.y + self.z*other.z
```

```python
class Polynomial():
    def __init__(self, data_init):
        if isinstance(data_init, dict):
            self._data=dict(data_init)
        else:
            self._data=list(data_init)
    
    def __setitem__(self, key, value):
        self._data[key] = value
    
    def __getitem__(self, key):
        if key in self._data:
            return self._data[key]
        else:
            return 0
    def __delitem__(self, key):
        self._data[key]=0
    
    
    def __add__(self, other):    
        pol = Polynomial({})
        for i in range(max(len(self._data), len(other._data))):
                if i<len(self._data) and i<len(other._data):
                    pol._data[i]=self._data[i] + other._data[i] 
                elif i>=len(self._data):
                    pol._data[i]=other._data[i]
                elif i>=len(other._data[i]):
                    pol._data[i]=self._data[i]  
        return pol
    
    def __sub__(self, other):
        pol = Polynomial({})
        for i in range(max(len(self._data), len(other._data))):
                if i<len(self._data) and i<len(other._data):
                    pol._data[i]=self._data[i] - other._data[i] 
                elif i>=len(self._data):
                    pol._data[i]=-other._data[i]
                elif i>=len(other._data[i]):
                    pol._data[i]=self._data[i]
        return pol
    
    def __neg__(self):
        pol = Polynomial({})
        for i in range(len(self._data)):
            pol._data[i] = -self._data[i]
        return pol
    def __mul__(self, other):
        pol=Polynomial({})
        for i in range(len(self._data)+len(other._data)-1):
            pol.append(0)
        for k in range(len(self._data)):
            for g in range(len(other._data)):
                pol[g+k]+=self._data[k]*other._data[g]
        return pol
    def __pow__(self, n):
        if n>1:
            p1 = Polynomial({})
            p2 = Polynomial({})
            p2._data = self._data
            for i in range(n-1):
                for i in range(len(self._data)+len(p2._data)):
                    p1._data[i] = 0
                for i in range(len(self._data)):
                    for j in range(len(p2._data)):
                        p1[i+j]+=self._data[i]*p2._data[j]
                p2._data = p1._data
                p1._data = {}
            return p2
        elif n==0:
            return 1
        elif n==1:
            return (self._data)
            
    
    def __str__(self):
        s = ' '
        for i in range(len(self._data)):
            if self._data[i] != 0:
                if i==0 :
                    s+=str(self._data[i])
                    
                elif i==1 and self._data[i]==1 and s ==' ' :
                    s+='x'
                elif i==1 and self._data[i]==1 and s !=' ' :
                    s+=' + x'
                elif i==1 and self._data[i]==-1 :
                    s+=' - x'
                
                elif i==1 and s ==' ' and self._data[i]>0 :
                    s+=str(self._data[i])+'*x'
                elif i==1 and s !=' ' and self._data[i]>0 :
                    s+=' + '+str(self._data[i])+'*x'
                elif i==1 and self._data[i]<0 :
                    s+=' - '+str(-self._data[i])+'*x'
                    
                elif self._data[i]==1 and s ==' ' :
                    s+='x^'+str(i)
                elif self._data[i]==1 and s !=' ' :
                    s+=' + x^'+str(i)
                elif self._data[i]==-1 :
                    s+=' - x^'+str(i)
                    
                elif self._data[i]>0 and s ==' ' :
                    s+=str(self._data[i])+'*x^'+str(i)
                elif self._data[i]>0 and s !=' ' :
                    s+=' + '+str(self._data[i])+'*x^'+str(i)
                elif self._data[i]<0 :
                    s+=' - '+str(-self._data[i])+'*x^'+str(i)
        return str(s)
```
